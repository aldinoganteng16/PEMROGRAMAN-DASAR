```python
#5. menggunakan for

batas_atas = 5
total_jumlah = 0

for i in range(1, batas_atas + 1):
    faktorial_sekarang = 1
    # Perulangan FOR bersarang
    for faktor in range(1, i + 1):
        faktorial_sekarang *= faktor

    total_jumlah += faktorial_sekarang

print(f"Total Jumlah Faktorial (1 hingga {batas_atas}) adalah: {total_jumlah}")
```

```python
#5. menggunakan while
target = 800000
mingguan = 50000
total = 0
minggu_ke = 0

while total < target:
    minggu_ke += 1
    total += mingguan

print(f"Dibuthkan {minggu_ke} minggu untuk mencapai target Rp {target:,}.")
```
